```python
def clean_files(df):
    # Convert "started_at" column to datetime format
    df['started_at'] = pd.to_datetime(df['started_at'])

    # Extract date and time components into separate columns
    df['start_date'] = df['started_at'].dt.date
    df['start_time'] = df['started_at'].dt.time

    # Drop the original "started_at" column if you don't need it anymore
    df.drop(columns=['started_at'], inplace=True)

    df['ended_at'] = pd.to_datetime(df['ended_at'])

    # Extract date and time components into separate columns
    df['end_date'] = df['ended_at'].dt.date
    df['end_time'] = df['ended_at'].dt.time

    # Drop the original "ended_at" column if you don't need it anymore
    df.drop(columns=['ended_at'], inplace=True)

    # Convert "start_time" and "end_time" columns to datetime format
    df['start_time'] = pd.to_datetime(df['start_time'], format='%H:%M:%S')
    df['end_time'] = pd.to_datetime(df['end_time'], format='%H:%M:%S')

    # Calculate the duration of the trip and store it in a new column
    df['trip_duration'] = df['end_time'] - df['start_time']

    # extract the duration in seconds, you can do the following:
    df['trip_duration_seconds'] = df['trip_duration'].dt.total_seconds()

    # extract the duration in minutes, you can do the following:
    df['trip_duration_minutes'] = df['trip_duration_seconds'] / 60

    #drop extra colummns
    df = df.drop(columns=["trip_duration", "trip_duration_seconds"])

    return df
```

```python
jan_2023_clean = clean_files(jan_2023)
jan_2023_clean.head()
```

```python
apr_2023_clean = clean_files(apr_2023)
apr_2023_clean.head()
```
